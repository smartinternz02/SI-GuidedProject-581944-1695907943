```python
import pandas as pd

file_name = 'data.csv'
df = pd.read_csv(file_name)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

plt.figure(figsize=(10, 6))
plt.hist(df['living area'], bins=20, color='skyblue')
plt.title('Distribution of Living Area')
plt.xlabel('Living Area')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['living area'], df['Price'], alpha=0.5, color='green')
plt.title('Scatter Plot: Price vs. Living Area')
plt.xlabel('Living Area')
plt.ylabel('Price')
plt.show()
```

```python
corr_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
import seaborn as sns
sns.pairplot(df[['living area', 'Price', 'number of bedrooms', 'condition of the house']])
plt.show()
```
